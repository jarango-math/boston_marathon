```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/llimllib/bostonmarathon/master/results/2012/results.csv")
```

```python
df['overall_pos'], df['overall_size'] = df['overall'].str.split('/', 1).str
```

```python
df['gender_pos'], df['gender_size'] = df['genderdiv'].str.split('/',1).str
df['division_pos'], df['division_size'] = df['division'].str.split('/',1).str
```

```python
df=df.drop(['division','overall','genderdiv'],axis=1)
```

```python
df.dtypes
```

```python
cols = ['official','overall_pos','overall_size','gender_pos','gender_size','division_pos','division_size']
df[cols]=df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes
```

```python
df=df.dropna(subset=['official'])
```

```python
df.isnull().sum()
```

```python
df['official_int']=df['official'].astype(int)
```

```python
df.head()
```

```python
plt.scatter(df['age'],df['official_int'])
plt.xlabel('age')
plt.ylabel('official time')
plt.show()
```

```python
from sklearn.cluster import estimate_bandwidth
bandwidth = estimate_bandwidth(df[['age','official_int']], quantile=0.2, n_samples=5000)
print(bandwidth)
```

```python
from sklearn.cluster import KMeans
```

```python
ap = KMeans(n_clusters=27)
y_pred = ap.fit_predict(df[['age','official']])
```

```python
plt.scatter(df['age'],df['official'], c=y_pred)
plt.xlabel('age')
plt.ylabel('official time')
plt.show()
```
